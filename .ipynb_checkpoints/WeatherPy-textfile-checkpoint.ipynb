```python
%matplotlib inline
import os
import json
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from citipy import citipy as cp
from config import api_key
from pandas.io.json import json_normalize

base_url='http://api.openweathermap.org/data/2.5/weather?'
params_OWM = {
    'units':'imperial',
    'APPID':api_key
}
```

```python
#Pull world data city list from text file onlocal machine.  Had to re-encode to UTF-8 in order for the file to be read properly
'''
citypath = os.path.join('resources','worldcitiespop1.txt')
cityread = pd.read_csv(citypath)
cities_df = pd.DataFrame(cityread, columns=['Latitude','Longitude'])
'''

citypath = os.path.join('city.list.json')
with open(citypath, encoding='utf8') as jsonfile:
    owmJSON = json.load(jsonfile)
cities_df = pd.DataFrame.from_dict(json_normalize(owmJSON), orient='columns')
del cities_df['country']
del cities_df['name']
cities_df.head()
```

```python
#Add columns to the dataframe to be populated later.  Rename certain fields from source pull on JSON load
cities_df['City'] = ""
cities_df['Country Code'] = ""
cities_df['Temperature (F)'] = ""
cities_df['Humidity (%)'] = ""
cities_df['Cloudiness (%)'] = ""
cities_df['Wind Speed (mph)'] = ""
cities_df.rename(columns={'coord.lat':'Latitude', 'coord.lon':'Longitude'}, inplace=True)

#Gather sample data for project, reset index
sampleSize = input('How large should the sample size be?  (Min. 500): ')
while int(sampleSize) < 500:
        sampleSize = input("Sample size to small.  Please choose a number of 500: ")
        sampleSize = int(sampleSize)
        if sampleSize == 500:
            break
            
cities_sample = cities_df.sample(sampleSize, replace=False)
cities_sample = cities_sample.reset_index(drop=True)
cities_sample.head()
```

```python


#Use CitiPy to find City and Country...
for index, row in cities_sample.iterrows():
    #get lat and lng
    cpLat = row['Latitude']
    cpLng = row['Longitude']
    city = cp.nearest_city(cpLat,cpLng)
    city.city_name
    cities_sample.loc[index,'City'] = (city.city_name).title()
    cities_sample.loc[index,'Country Code'] = city.country_code
    
    #print(temp_lat)
#     result = cp.nearest_city(cities_sample['Latitude'],cities_sample['Longitude'])
#     print(result)
#     city['CP City'] = result.city_name
#     city['CP Country'] = result.country_code

#     owmQuery = row['City'] + ',' + row['Country Code']
    owmQuery = row['id']
    params_OWM['id'] = owmQuery
#     params_OWM['q']=owmQuery
    #print(owmQuery)
    owmResponse = requests.get(base_url, params_OWM).json()
    #print(json.dumps(owmResponse, indent=4))
    try:
        cities_sample.loc[index, 'Temperature (F)'] = owmResponse['main']['temp_max']
        cities_sample.loc[index, 'Humidity (%)'] = owmResponse['main']['humidity']
        cities_sample.loc[index, 'Cloudiness (%)'] = owmResponse['clouds']['all']
        cities_sample.loc[index, 'Wind Speed (mph)'] = owmResponse['wind']['speed']
    except (KeyError, IndexError):
        print(f'I am Bender, please insert girder.  Something is missing...skipping...')

cities_sample.dropna(inplace=True)

#    api.openweathermap.org/data/2.5/weather?q={city name},{country code}&units=imperial
# '''
# main
# +main.temp
# +main.humidity
# wind
# +windspeed
# clouds
# +clouds.all
# '''
    
```

```python
lat = cities_sample['Latitude']
temp = cities_sample['Temperature (F)']
humid = cities_sample['Humidity (%)']
cloud = cities_sample['Cloudiness (%)']
wind = cities_sample['Wind Speed (mph)']
```

```python
plt.scatter(lat, temp, c='lightcoral', alpha=0.7, label='Temp (F)', edgecolors='firebrick', linewidths=2)
plt.title('Temperature (F) vs. Latitude')
plt.xlabel('Latitude')
plt.ylabel('Temperature (F)')
plt.xlim(-70,70)
plt.grid(alpha=0.3)
```

```python
plt.scatter(lat, humid, c='lawngreen', alpha=0.7, label='Humidity', edgecolors='darkolivegreen', linewidths=2)
plt.title('Humidity (%) vs. Latitude')
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')
plt.xlim(-70,70)
plt.grid(alpha=0.3)
```

```python
plt.scatter(lat, cloud, c='skyblue', alpha=0.7, label='Cloudiness', edgecolors='dodgerblue', linewidths=2)
plt.title('Cloudiness (%) vs. Latitude')
plt.xlabel('Latitude')
plt.ylabel('Cloudiness (%) vs. Latitude')
plt.grid(alpha=0.3)
plt.xlim(-70,70)
plt.show()
```

```python
plt.scatter(lat, wind, c='violet', alpha=0.7, label='Wind Speed', edgecolors='darkviolet', linewidths=2)
plt.title('Wind Speed (mph) vs. Latitude')
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (mph)')
plt.grid(alpha=0.3)
plt.xlim(-70,70)
plt.show()
```
